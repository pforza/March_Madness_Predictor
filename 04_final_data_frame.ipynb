```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
team_data_clean = pd.read_csv("team_data_clean.csv")
season_stats= pd.read_csv("season_stats.csv")
```

```python
team_data_clean = team_data_clean.drop("Unnamed: 0", axis=1)
team_data_clean
```

```python
season_stats= season_stats.drop("Unnamed: 0", axis=1)
season_stats
```

```python
df = team_data_clean.merge(season_stats, how='inner', on=['TeamID', 'Season'])
df
```
